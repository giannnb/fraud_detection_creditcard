**Question 1:** How many customers are in the file?
(as help, we have added some comments and starter code to help you structure the solution)

```python
import pandas as pd

path = '/content/sample_data/customers.csv'

df = pd.read_csv(path)

# Cantidad de Registros Totales de Clientes
total_rows = df['CustomerID'].count()
print(f"Cantidad total de registros de clientes: {total_rows}")

print("=" * 60)

# Cantidad de Clientes Unicos (Respuesta final a la pregunta)
unique_customers = df['CustomerID'].nunique()
print(f"Cantidad de clientes (unicos): {unique_customers}")
```

**Question 2:** In how many different states do the customers live in?

```python
import pandas as pd

path = '/content/sample_data/customers.csv'

df = pd.read_csv(path)

# Limpio para evitar duplicados erroneos
df['State'] = df['State'].str.strip().str.upper()

# Cantidad de estados diferentes
diferent_states = df['State'].nunique()
print(f"Cantidad de estados diferentes: {diferent_states}")

print("=" * 60)

list_states = df['State'].unique()
print(f"Lista de estados unicos: {list(list_states)}")
```

**Question 3** What is the state with most customers?

```python
import pandas as pd

path = '/content/sample_data/customers.csv'

df = pd.read_csv(path)

df['State'] = df['State'].str.strip().str.upper()

# Conteo de Valores por Estado
state_counts = df['State'].value_counts()
print(f"Conteo de clientes por estado: {state_counts}")

# Estado con mas Clientes
top_state_name = state_counts.idxmax()
top_state_count = state_counts.max()

print("=" * 60)
print(f"Estado con mas clientes: {top_state_name} (Total: {top_state_count})")
```

**Question 4** What is the state with the least customers?

```python
import pandas as pd

path = '/content/sample_data/customers.csv'

df = pd.read_csv(path)

df['State'] = df['State'].str.strip().str.upper()

# Conteo de Valores por Estado
state_counts = df['State'].value_counts()
print(f"Conteo de clientes por estado: {state_counts}")

# Estados con menos clientes
least_states = state_counts[state_counts == state_counts.min()]
print("=" * 60)
for estado, cantidad in least_states.items():
    print(f"- {estado}: {cantidad}")
```

**Question 5:** What is the most common last name?

```python
import pandas as pd

path = '/content/sample_data/customers.csv'

df = pd.read_csv(path)

# Apellido que mas se repite
last_name = df['LastName'].value_counts().idxmax()
quantity = df['LastName'].value_counts().max()

print(f"Apellido que mas se repite: {last_name} (Total: {quantity})")

print("=" * 60)

group_by_last_name = df.groupby('LastName')['CustomerID'].nunique()

duplicados = group_by_last_name[group_by_last_name > 1]

if duplicados.empty:
  print("No hay apellidos comunes entre usuarios unicos")
else:
  last_name_common = duplicados.idxmax()
  quantity_common = duplicados.max()
  print(f"Apellido mas comun entre usuarios unicos: {last_name_common} (Total: {quantity_common})")
```

**Question #1:** How many unique orders are in the orders.csv file?

**Question #2:** What is the average number of items per order (rounded to two decimal places)?

**Question #3:** What is the highest number of items per order?

**Question #4:** What is the number of orders placed in October 2021?

**Question #5:** Which customer spent the most amount of money in 2021?

**Question #6:** Historically, what is the best month for sales?

Once you get your answers, remember to go back to the course and introduce them in the multiple choice quiz

```python
from numpy import average
import pandas as pd

path = '/content/sample_data/orders.csv'

df = pd.read_csv(path)

# Shape
print(df.shape)

print("=" * 60)

# Question 1: Cantidad de ordenes unicas
unique_orders = df['OrderID'].nunique()
print(f"Cantidad de ordenes unicas: {unique_orders}")

print("=" * 60)

# Question 2: Cantidad promedio de items por pedido
items_per_order = df.groupby('OrderID').size()
average_items_per_order = items_per_order.mean()
final_result = round(average_items_per_order, 2)
print(f"Cantidad promedio de items por pedido: {final_result}")

print("=" * 60)


# Question 3: Numero maximo de items por pedido
max_items_per_order = items_per_order.max()
print(f"Numero maximo de items por pedido: {max_items_per_order}")

print("=" * 60)

# Question 4: Número de pedidos realizados en octubre de 2021
df['Date'] = pd.to_datetime(df['Date'])
fecha_inicio = '2021-10-01'
fecha_fin = '2021-10-31'
filter_by_date = df[(df['Date'] >= fecha_inicio) & (df['Date'] <= fecha_fin)]
orders_in_october = filter_by_date['OrderID'].nunique()
print(f"Numero de pedidos realizados en octubre de 2021: {orders_in_october}")

print("=" * 60)


# Question 5: Cliente que mas gasto en 2021
df['Date'] = pd.to_datetime(df['Date'])
fecha_de_inicio = '2021-01-01'
fecha_de_fin = '2021-12-31'

filter_by_date = df[(df['Date'] >= fecha_de_inicio) & (df['Date'] <= fecha_de_fin)]
unique_orders_2021 = filter_by_date.drop_duplicates(subset='OrderID')
spent_by_customer = unique_orders_2021.groupby('CustomerID')['OrderTotal'].sum().sort_values(ascending=False)

customer_most_amount = spent_by_customer.idxmax()
total_amount = spent_by_customer.max()

print(f"Cliente que mas gasto en 2021: {customer_most_amount} (Total: {total_amount})")


print("=" * 60)

# Question 6: Mejor mes para ventas
group_by_period = df.groupby(df['Date'].dt.to_period("M"))['OrderTotal'].sum()
best_month = group_by_period.idxmax()
best_month_total = group_by_period.max()
print(f"El mejor mes histórico de ventas fue {best_month}, con {best_month_total} ventas")
```
